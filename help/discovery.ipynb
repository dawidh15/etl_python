```python
import pandas as pd

# Example DataFrame
df = pd.DataFrame({
    "name": ["Alice", "Bob", "Charlie"],
    "age": [25, 30, 35],
    "city": ["NY", "LA", "Chicago"]
})

# Columns you want to combine
cols = ["name", "age", "city"]

# Convert selected columns to string and join row-wise
df["combined"] = df[cols].astype(str).agg(" ".join, axis=1)

print(df)

df["combined"] = df[cols].astype(str).replace("nan", "").agg(" ".join, axis=1)

print(df)
```

```python
import pandas as pd
from eccl.Table import TableMetadata, Table
from eccl.Combiner import PandasCombiner

data = pd.read_csv("data/sample_data.csv")

#read Metadata
metadata = TableMetadata(json_path="data/table_1_definition.json",schema_path="schema/table_schema.json")

# Create table object with metadata
myTable = Table(metadata=metadata)

#Initiate instance combiner
inst_combiner = PandasCombiner()

myTable.set_combiner(new_combiner=inst_combiner)

# Create PKs

# Combine type1
myTable.use_combiner(data=data,pk=None, cols_to_combine=myTable.table_metadata.type1)

```
